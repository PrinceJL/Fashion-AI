```python
import os
import json
import numpy as np
import pandas as pd
from PIL import Image
from sklearn.model_selection import train_test_split
from torchvision import transforms
import torch
from torch.utils.data import Dataset, DataLoader
import json
from IPython.display import display, HTML
import matplotlib.pyplot as plt
```

```python
# Use the absolute path that we know works
base_path = "../DeepFashionData"
img_dir = os.path.join(base_path, "images")
text_desc_path = os.path.join(base_path, "captions.json")
shape_label_path = os.path.join(base_path, "labels", "shape", "shape_anno_all.txt")
fabric_label_path = os.path.join(base_path, "labels", "texture", "fabric_ann.txt")
color_label_path = os.path.join(base_path, "labels", "texture", "pattern_ann.txt")
segm_dir = os.path.join(base_path, "segm")

# Function to check if all paths exist
def check_paths():
    paths = {
        "Images Directory": img_dir,
        "Captions File": text_desc_path,
        "Shape Labels": shape_label_path,
        "Fabric Labels": fabric_label_path,
        "Color Labels": color_label_path,
        "Segmentation Directory": segm_dir
    }
    
    for name, path in paths.items():
        if os.path.exists(path):
            print(f"✓ {name} exists: {path}")
        else:
            print(f"✗ {name} does not exist: {path}")

# Check all paths before proceeding
check_paths()
```

```python
image_files = sorted(os.listdir(img_dir))[:10]

# Create a DataFrame to store image information
image_data = []

# Load captions if the file exists
captions = {}
if os.path.exists(text_desc_path):
    with open(text_desc_path, 'r') as f:
        captions = json.load(f)

# Create a figure with a grid of subplots for the images
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
axes = axes.flatten()

# Process each image
for i, img_file in enumerate(image_files):
    img_path = os.path.join(img_dir, img_file)
    
    # Get image caption if available
    caption = captions.get(img_file, "No caption available")
    
    # Load and display the image
    try:
        img = Image.open(img_path)
        axes[i].imshow(img)
        axes[i].set_title(f"Image {i+1}: {img_file[:15]}...")
        axes[i].axis('off')
        
        # Add to our data collection
        image_data.append({
            'Image': img_file,
            'Size': f"{img.width}x{img.height}",
            'Caption': caption[:50] + "..." if len(caption) > 50 else caption
        })
    except Exception as e:
        axes[i].text(0.5, 0.5, f"Error loading image: {str(e)}", 
                    ha='center', va='center', color='red')
        axes[i].axis('off')
        
        image_data.append({
            'Image': img_file,
            'Size': "Error",
            'Caption': "Error loading image"
        })

plt.tight_layout()
plt.show()

# Create and display the table with image information
df = pd.DataFrame(image_data)
display(HTML(df.to_html(index=False)))

print(f"Successfully displayed {len(image_files)} images from the DeepFashion dataset")
```

```python
# Define preprocessing function
def preprocess_image_and_mask(image, mask, image_size=(256, 256)):
    # Define transformations
    image_transform = transforms.Compose([
        transforms.Resize(image_size),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalize to [-1, 1]
    ])

    mask_transform = transforms.Compose([
        transforms.Resize(image_size, interpolation=Image.NEAREST),
        transforms.ToTensor()
    ])

    image_tensor = image_transform(image)
    mask_tensor = mask_transform(mask)

    # Optionally convert mask to binary (if needed)
    mask_tensor = (mask_tensor > 0.5).float()  # Convert to binary mask if needed

    return image_tensor, mask_tensor

def visualize_image_and_mask(image, mask):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    axes[0].imshow(image)
    axes[0].set_title("Image")
    axes[0].axis("off")

    axes[1].imshow(mask, cmap="gray")
    axes[1].set_title("Mask")
    axes[1].axis("off")

    plt.tight_layout()
    plt.show()
```

```python
# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
# Function to get and match image and segmentation files
def get_matched_image_and_mask_files():
    image_files = sorted(os.listdir(img_dir))  # Get sorted list of image files
    mask_files = sorted(os.listdir(segm_dir))  # Get sorted list of mask files

    # Create a dictionary to map each image file to its corresponding mask file
    image_mask_mapping = {}

    for image_filename in image_files:
        # Assuming the mask filename is the same as the image filename with a different extension
        image_name, _ = os.path.splitext(image_filename)
        
        # Find the corresponding mask file
        mask_filename = None
        for mask in mask_files:
            if image_name in mask:
                mask_filename = mask
                break
        
        if mask_filename:
            image_mask_mapping[image_filename] = mask_filename

    return image_mask_mapping

# Function to load an image and its corresponding segmentation mask
def load_image_and_mask(image_filename, mask_filename):
    # Load the image (RGB) and mask (grayscale)
    image_path = os.path.join(img_dir, image_filename)
    mask_path = os.path.join(segm_dir, mask_filename)
    
    image = Image.open(image_path).convert("RGB")
    mask = Image.open(mask_path).convert("L")  # Convert mask to grayscale (binary)
    
    return image, mask

# Example of using the matched files
def process_matched_files():
    image_mask_mapping = get_matched_image_and_mask_files()

    # Initialize a list to store the processed data (for further use or model training)
    processed_data = []

    # Get the total number of files to process
    total_files = len(image_mask_mapping)
    
    # Loop through the matched files
    for idx, (image_filename, mask_filename) in enumerate(image_mask_mapping.items(), start=1):
        # Load image and mask
        image, mask = load_image_and_mask(image_filename, mask_filename)
        
        # Preprocess the image and mask
        image_tensor, mask_tensor = preprocess_image_and_mask(image, mask)
        
        # Store the processed data (image_tensor, mask_tensor)
        processed_data.append((image_tensor, mask_tensor))
        
        # Print progress for every 10% of the files processed
        if idx % (total_files // 10) == 0 or idx == total_files:
            print(f"Processed {idx}/{total_files} files ({(idx / total_files) * 100:.2f}%)")
        
        # For debugging purposes, you can visualize the processed data
        if len(processed_data) == 1:  # Only visualize the first item
            visualize_image_and_mask(image, mask)
        
    return processed_data

# Run the process and get the matched files
processed_data = process_matched_files()

# Example of accessing the preprocessed data
print(f"Processed {len(processed_data)} image-mask pairs")
```

```python
import torch

# Load the file
processed_data = torch.load("output/processed_data.pt")

print(f"Loaded {len(processed_data)} image-mask pairs from processed_data.pt")
```

```python
from torch.utils.data import Dataset

class SegmentationDataset(Dataset):
    def __init__(self, data):
        self.data = data  # List of (image_tensor, mask_tensor)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        image_tensor, mask_tensor = self.data[idx]
        return image_tensor, mask_tensor
```

```python
from torch.utils.data import DataLoader, random_split, TensorDataset

dataset = SegmentationDataset(processed_data)
train_loader = DataLoader(dataset, batch_size=8, shuffle=True)

# Assume processed_data = [(image_tensor, mask_tensor), ...]
total_size = len(processed_data)
train_size = int(0.9 * total_size)
test_size = total_size - train_size

# Split processed data
train_data, test_data = random_split(processed_data, [train_size, test_size])

# Create DataLoaders
train_loader = DataLoader(train_data, batch_size=8, shuffle=True)
test_loader = DataLoader(test_data, batch_size=8, shuffle=False)

print(f"Train size: {len(train_data)}")
print(f"Test size: {len(test_data)}")
```

```python
import segmentation_models_pytorch as smp
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = smp.Unet(
    encoder_name="resnet34",    # Or try 'mobilenet_v2'
    encoder_weights="imagenet", # Pretrained on ImageNet
    in_channels=3,              # RGB images
    classes=1,                  # Binary segmentation (1 class + background)
).to(device)
```

```python
import torch.nn as nn
import torch.optim as optim

loss_fn = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
```

```python
def compute_accuracy(preds, masks):
    preds = torch.argmax(preds, dim=1)  # If multi-class
    correct = (preds == masks).float()
    accuracy = correct.sum() / correct.numel()
    return accuracy.item()
    
import glob

def load_latest_checkpoint(model, folder="checkpoints"):
    checkpoints = glob.glob(os.path.join(folder, "*.pth"))
    if not checkpoints:
        print("⚠️ No checkpoints found. Starting from scratch.")
        return model, 0  # No checkpoint
    latest = max(checkpoints, key=os.path.getctime)
    model.load_state_dict(torch.load(latest))
    print(f"🔁 Loaded checkpoint: {latest}")
    # Get the starting epoch from the filename
    epoch_start = int(os.path.basename(latest).split("_")[1].split(".")[0])
    return model, epoch_start
```

```python
model, start_epoch = load_latest_checkpoint(model)
model.to(device)

num_epochs = 10

model.train()
for epoch in range(start_epoch, num_epochs):
    epoch_loss = 0
    epoch_acc = 0
    print(f"\nEpoch {epoch+1}/{num_epochs} -------------------------")
    
    for batch_idx, (images, masks) in enumerate(train_loader):
        images, masks = images.to(device), masks.to(device)

        preds = model(images)
        loss = loss_fn(preds, masks)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        acc = compute_accuracy(preds, masks)
        epoch_acc += acc
        epoch_loss += loss.item()

        print(f"  Batch {batch_idx+1}/{len(train_loader)} | Loss: {loss.item():.4f} | Acc: {acc:.4f}")

    avg_loss = epoch_loss / len(train_loader)
    avg_acc = epoch_acc / len(train_loader)
    print(f"==> Epoch {epoch+1} Completed | Avg Loss: {avg_loss:.4f} | Avg Acc: {avg_acc:.4f}")

    # Save model
    checkpoint_path = f"checkpoints/epoch_{epoch+1}.pth"
    os.makedirs("checkpoints", exist_ok=True)
    torch.save(model.state_dict(), checkpoint_path)
    print(f"✅ Saved checkpoint to {checkpoint_path}")
```

```python
model.eval()
total, correct = 0, 0

with torch.no_grad():
    for images, masks in test_loader:
        images, masks = images.to(device), masks.to(device)
        preds = model(images)

        # Apply threshold or argmax depending on output
        predicted = (preds > 0.5).float()  # for binary seg
        correct += (predicted == masks).sum().item()
        total += masks.numel()

    accuracy = correct / total * 100
    print(f"🧪 Test Accuracy: {accuracy:.2f}%")
```

```python
def visualize_prediction(image_tensor, true_mask, pred_mask):
    image_np = image_tensor.squeeze().cpu().permute(1, 2, 0).numpy()
    true_mask_np = true_mask.squeeze().cpu().numpy()
    pred_mask_np = pred_mask.squeeze().cpu().numpy()

    fig, axs = plt.subplots(1, 3, figsize=(12, 4))
    axs[0].imshow(image_np)
    axs[0].set_title("Image")
    axs[1].imshow(true_mask_np, cmap="gray")
    axs[1].set_title("Ground Truth")
    axs[2].imshow(pred_mask_np, cmap="gray")
    axs[2].set_title("Prediction")
    plt.show()
```
